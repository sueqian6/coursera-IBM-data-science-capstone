```
r = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(r.content)

codes = []
boroughs = []
neighborhoods = []

j = 1
for td in soup.find_all("table")[0].find_all("td"):
    if (j % 3 == 1):
        codes.append(td.text)
        j = j + 1
    elif (j % 3 == 2):
        boroughs.append(td.text.replace("]","").replace("\n",""))
        j = j + 1
    else:
        neighborhoods.append(td.text.replace("]","").replace("\n",""))
        j = j + 1

# Create the Pandas dataframe
column_names = ["PostalCode","Borough","Neighborhood"]
df_tor = pd.DataFrame(list(zip(codes, boroughs, neighborhoods)), columns = column_names)
# Drop any rows which do not have an assigned borough
df_tor = df_tor[df_tor["Borough"] != "Not assigned"]
# Replace any unassigned neighborhoods with the name of the borough
df_tor["Neighborhood"] = df_tor["Neighborhood"].mask(df_tor["Neighborhood"].apply(lambda x: x.upper() == "NOT ASSIGNED"), df_tor["Borough"], axis = 0)
df_tor["Neighborhood"] = df_tor.groupby(["PostalCode"])["Neighborhood"].transform(lambda x: ", ".join(x))
df_tor.drop_duplicates("PostalCode",inplace = True)
df_tor.reset_index(drop = True, inplace = True)
df_tor
```
